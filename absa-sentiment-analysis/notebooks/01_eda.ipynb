# EDA: Aspect-Based Sentiment Analysis Dataset

Load processed splits (or raw XML), inspect class distribution, sentence length, and aspect terms.

```python
import sys
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT))

from src.data_loader import load_processed_splits
from src.preprocess import LABEL2ID
```

```python
processed_dir = ROOT / "data" / "processed"
if (processed_dir / "train.csv").exists():
    train_df, val_df, test_df = load_processed_splits(str(processed_dir))
    train_df["split"] = "train"
    val_df["split"] = "val"
    test_df["split"] = "test"
    df = pd.concat([train_df, val_df, test_df], ignore_index=True)
else:
    df = pd.DataFrame({
        "sentence": ["Great food but slow service."] * 30 + ["The ambiance was nice."] * 20,
        "aspect_term": ["food", "service", "ambiance"] * 17[:50],
        "polarity": ["positive"] * 20 + ["negative"] * 15 + ["neutral"] * 15,
    })
    df["split"] = "train"
df.head(10)
```

```python
# Class distribution
fig, ax = plt.subplots(figsize=(8, 4))
df["polarity"].value_counts().plot(kind="bar", ax=ax, color="steelblue", edgecolor="black")
ax.set_title("Polarity distribution")
ax.set_xlabel("Polarity")
ax.set_ylabel("Count")
plt.tight_layout()
(ROOT / "results" / "figures").mkdir(parents=True, exist_ok=True)
plt.savefig(ROOT / "results" / "figures" / "eda_polarity_dist.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# Sentence length (word count)
df["sent_len"] = df["sentence"].str.split().str.len()
fig, ax = plt.subplots(figsize=(8, 4))
df["sent_len"].hist(bins=30, ax=ax, color="coral", edgecolor="black")
ax.set_title("Sentence length (words)")
ax.set_xlabel("Length")
ax.set_ylabel("Count")
plt.tight_layout()
plt.savefig(ROOT / "results" / "figures" / "eda_sent_len.png", dpi=150, bbox_inches="tight")
plt.show()
```
